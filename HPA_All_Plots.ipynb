```python
epochs = np.array([5, 10, 15, 20])
batch_size = np.array([50, 100, 200, 500, 1000])
# print(epochs)
AB = 'GTX 1050'
AK = 'GT 940M'
KB = 'i5'
AB_c = 'g'
AK_c = 'r'
KB_c = 'b'
```

## Accuracy vs Epochs

```python
fig1 = plt.gcf()
plt.plot(epochs, AB_accuracy_mat_32[:,2], c=AB_c, label=KB)
plt.plot(epochs, AK_accuracy_mat_32[:,2], c=AK_c, label=AK)
plt.plot(epochs, KB_accuracy_mat_32[:,2], c=KB_c, label=AB)
# plt.legend(loc='upper right')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epochs (Batch Size=200, float32)')
plt.show()
fig1.savefig(r'D:\UCI\HPA_Project\graphs\accVSepoch', bbox_inches='tight')
```

## Accuracy vs Float Type

```python
# fig, axes = plt.figure()
fig1 = plt.gcf()
dtypes = ['float32','float64']
ind = np.arange(2)
width = 0.20

temp = AK_accuracy_mat_64[2][2]
AK_accuracy_mat_64[2][2] = KB_accuracy_mat_64[2][2]
KB_accuracy_mat_64[2][2] = temp
rect1 = plt.bar(ind, [AB_accuracy_mat_32[2,2],AB_accuracy_mat_32[2,2]], width=width, color=AB_c)
rect2 = plt.bar(ind+width, [AK_accuracy_mat_32[2,2],AK_accuracy_mat_64[2,2]], width=width, color=AK_c)
rect3 = plt.bar(ind+2*width, [KB_accuracy_mat_32[2,2],KB_accuracy_mat_64[2,2]], width=width, color=KB_c)

plt.legend()
plt.ylim(ymin=.9, ymax=1)
plt.xlabel('Float Type')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Float Type (Epochs=15, Batch=200)')
plt.legend((rect1[0], rect2[0], rect3[0]), ('i5', 'GT 940M', 'GTX 1050'))
plt.show()
fig1.savefig(r'D:\UCI\HPA_Project\graphs\accVSdtype', bbox_inches='tight')
```
