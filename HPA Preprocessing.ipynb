```python
processed_flag = False

# remove intermittent headers from line
filename = r'D:\UCI\HPA_Project\Akshay data\nvidia.txt'
header = 'date time gpu pwr temp sm mem enc dec mclk pclk fb bar1 rxpci txpci\n'
with open(filename, "r") as f:
    lines = f.readlines()
    
with open(filename, "r") as f:
    line1 = f.readline()
    if line1==header:
        processed_flag=True
print(processed_flag)

if not processed_flag:
    with open(filename, "w") as f:
        f.write(header)
        for line in lines:
            if line[0] != '#':
                f.write(line)
```

```python
epoch = datetime.datetime.utcfromtimestamp(0)

def unix_time_secs(dt):
    return round((dt - epoch).total_seconds())

def convert_to_timesince_epoch(df):
    date=str(df['date'])
    time=df['time']
    yyyy, MM, dd = date[:4], date[4:6], date[6:]
    hh, mm, ss = time.split(':')
    return unix_time_secs(datetime.datetime(int(yyyy), int(MM), int(dd), int(hh), int(mm), int(ss)))
```

```python
df = pd.read_csv(filename, delim_whitespace=True)
df.drop(['gpu','pwr','enc','dec'], axis=1, inplace=True)  # df.columns is zero-based pd.Index
```

```python
df['datetime']=df.apply(lambda row: convert_to_timesince_epoch(row), axis=1)
```

```python
df.drop(['date','time'], inplace=True, axis=1)
df.head()
```

```python
df.describe()
```

```python
data = df.values

def find_mean(stime, etime):
    sindex, eindex = -1, -1
    while sindex==-1 or eindex==-1:
        for i in range(data.shape[0]):
            if stime==data[i][-1]:
                sindex=i
            if etime==data[i][-1]:
                eindex=i
        stime += 1
        etime += 1
    
    if sindex>eindex:
        print('There seems to be some problem...')
        print('Start time={}, End time={}'.format(stime, etime))
        print('Start index={}, End index={}'.format(sindex, eindex))
        return find_mean(stime, etime+1)
        
    means = np.mean(data[sindex:eindex+1], axis=0)
    means = means[:-1] # remove the mean of timestamp
    return means
   
```

```python
filename = r'D:\UCI\HPA_Project\Akshay data\f64.csv'
mat64 = np.zeros([13,10])
with open(filename, 'r') as f:
    i = 0
    for line in f:
        epoch, stime, etime, duration = line.split(',')
        mean_vals = find_mean(int(stime), int(etime))
        mean_vals = np.append(mean_vals, int(duration)) # append the duration per batch size to the data
        mat64[i] = mean_vals
        i += 1
```
